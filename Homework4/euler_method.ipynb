```python
import numpy as np
import plotly
import math
import scipy
import matplotlib.pyplot as plt
```

```python
def dx_dt(t, x, y, z, sigma, b, r):
    return sigma * (y - x)

def dy_dt(t, x, y, z, sigma, b, r):
    return x * (r - z) - y

def dz_dt(t, x, y, z, sigma, b, r):
    return x * y - b * z
```

```python
sigma = 10
b = 8/3
rs = [0.5, 10, 13.927, 17, 24.06, 30]
dt = 0.0001
max_t = 100
```

```python
def solve_euler(r, t0=0, x0=13.41265629, y0=13.46430003, z0=33.46156416):
    t = t0
    x = x0
    y = y0
    z = z0
    
    ts = [t0]
    xs = [x0]
    ys = [y0]
    zs = [z0]
    
    while t < max_t:
        t += dt
        x += dx_dt(t, x, y, z, sigma, b, r) * dt
        y += dy_dt(t, x, y, z, sigma, b, r) * dt
        z += dz_dt(t, x, y, z, sigma, b, r) * dt
        
        ts.append(t)
        xs.append(x)
        ys.append(y)
        zs.append(z)
    
    return ts, xs, ys, zs
```

```python
for r in rs:
    ts, xs, ys, zs = solve_euler(r)
    
    print('r = {0}'.format(r))
    plt.plot(ts, xs, 'red')
    plt.show()
    plt.plot(ts, ys, 'green')
    plt.show()
    plt.plot(ts, zs, 'blue')
    plt.show()
```
